```python
import tensorflow as tf
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
temp_df = pd.read_csv('/Celsius+to+Fahrenheit.csv')
temp_df.head(5)
temp_df.tail(10)
```

```python
temp_df.info()
temp_df.describe().info()
```

```python
sns.scatterplot(temp_df['Celsius'], temp_df['Fahrenheit'])
```

```python
X_train = temp_df['Celsius']
y_train = temp_df['Fahrenheit']
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(units = 3, input_shape = [1]))
```

```python
model.summary()
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam(1), loss = 'mean_squared_error')
```

```python
epochs_hist = model.fit(X_train, y_train, epochs = 50)
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.title('Model Loss Progress During Training')
plt.xlabel('Epochs')
plt.ylabel('Training Loss')
plt.legend('Training Loss')
```

```python
model.get_weights()
```

```python
Temp_C = 50
Temp_f = model.predict([Temp_C])
print("temperature in DegF using Trained ANN: ", Temp_f)
```

```python
Temp_f = 9/5 * Temp_C + 32
print("temperature in DegF using equation ANN: ", Temp_f)
```
